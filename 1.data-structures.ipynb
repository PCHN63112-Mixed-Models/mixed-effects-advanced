# Data Structures
To begin, we will start with a general overview of different data structures that mixed-effects models can be applied to. This particular taxonomy is taken from [West, Welch & Galecki (2022)](https://www.taylorfrancis.com/books/mono/10.1201/9781003181064/linear-mixed-models-brady-west-kathleen-welch-andrzej-galecki?context=ubx&refId=4dbda113-7f59-4ba9-a710-897ef536cfee), which is the recommended text for this part of the unit. In general, most datasets you will encounter will fit into one of these categories and so it is useful to understand both the similarities and differences between them. Remember as well that the model knows nothing of different data structures. The whole point of this is to understand how we can represent these structures in a sensible way using a mixed-effects framework. However, the model does not know how *subjects* are different from *litters*, or how *classes* are different from *schools*. So, even if your data does not neatly fit into one of these definitions, remember that these are *means to an end*, rather than *rules*. Once you understand the process of translating from *structure* to *model*, you should be able to work with any complex structure, using one of these examples as a starting point.

## Types of Structure
We will now briefly explore each of the structures we will use in this lesson: *clustered*, *repeated measurements*, *longitudinal* and *clustered longitudinal*.

### Clustered
The first type of data structure is referred to as *clustered* data. Unlike all the examples we have seen so far, clustered data has no repeated measurements on the units of analysis. Instead, the units are grouped, or nested, within *clusters* of units. 

For example, the first clustered dataset we will see concerns the birth weight of rat pups. The rat pups represent the units of analysis and each only has a single weight, so there are no repeated measurements here. However, the pups can be grouped into *litters* from the same mother. The assumption here is that pups from the same litter may be correlated in some fashion due to shared genetics and rearing environment. Perhaps one mother consistently has pups that are *underweight*, or *overweight*? We can also think of this as *two levels* of variation, one between the weights within a litter and one between litters.

Similarly, our second clustered dataset concerns maths scores from students. Each student is the unit of analysis and only has a single maths score, so again there are no repeated measurements. Nevertheless, there is structure here, as each student is clustered within a class, and classes are clustered within a school. Much like the rat pups, the assumptions here is that students from the same class may be correlated, but also that classes from the same school may also be correlated. Perhaps some schools generally produce higher maths grades than other schools, and within those schools some classes also perform consistently better or worse. There are therefore *three levels* of variation here: scores, classes and schools.

### Repeated Measurements
The second type of data structure is *repeated measurements*. Obviously, we know a fair amount about this type of data by this point. However, as has been mentioned multiple times so far, there is a key difference between repeated measurements *with* replications and repeated measurements *without* replications. The power of LME models is only really available when we have replications. In other words, when there are *multiple* observations per-subject and per-repeat. Without this, we are confined to *random-intercepts* and must assume a compound symmetric covariance structure. This is effectively a repeated measures ANOVA without any sphericity correction.

For example, the repeated measurements dataset we will use as an example this week concerns measurements of a chemical in the brains of rats under two different treatments. The treatments are defined *within-subject*, so each rat (the unit of analysis) gets *both*. Importantly, these measurements are collected from *three* different brain regions. Each rat therefore has 6 measurements, 3 for each treatment. If our interest is only on the effects of treatment, we can think of these as 3 replications per-repeat. Rather than averaging-over these replications to create a mean per-rat and per-treatment, we can instead use the replications directly in the model. This then allows much greater scope to capture complex covariance structures.

### Longitudinal
The third type of data structure is *longitudinal*. This is very similar to repeated measurements in the sense that each unit of analysis is measured *multiple times*. The difference is that these measurements are both *spaced* over a longer period of time and have a *natural order*. Repeated measurements are typically collected close together in time and the order of collection is largely immaterial. For instance, in the rat brain example above, the measurements were taken from each brain at the same point in time and it does not matter what order the brain regions were sampled in. Typically, this means that missing data is *less* common under repeated measurements, unless there is equipment failure or a subject decided to withdraw midway through collection[^ratbrain-foot]. 

*Longitudinal* data is spaced much further in time, typically days, weeks, months or years. This has *two* consequences. Firstly, there is a naturalistic ordering to the measurements that needs to be taken into account. If we are looking for patterns in the data, we want these to be defined in terms of *time*. Secondly, the gap of time between the measurements needs to be considered. If we take measurements at 1 year, 2 years and 5 years, we cannot treat the effect of a 2 year gap the same as a 5 year gap. The amount of time passed needs to be in the model. This is not true of repeated measurements, where we can model the effects of the different levels without consideration of *order* or *measurement gap*. Missing data is also much more common in longitudinal data, as you might imagine. This is where mixed-effects models have an advantage over other methods, thanks to *pooling* and *shrinkage*.

As an example of longitudinal data, we will examine a dataset of autistic children measured on a scale of *socialisation* at ages 2, 3, 5, 9 and 13 years. Notice that there are uneven gaps between these measurements, which needs to be accommodated. In addition, there will almost certainly be missing data considering the practical difficulties of getting children to reliably return for measurement across 11 years of their lives. Of importance is that we need to consider the *reasons* for missing data. If these are innocuous, such as the family moving house, then this is unfortunate be not disastrous. However, if the missingness speaks to some sort of *confound* then we are in trouble. For instance, if the reason why children do not return is *related* to poor socialisation skills, then this will create a bias. For instance, if the *worst* cases disappear over time, it will appear as if average socialisation is improving, when the *opposite* might actually be true.

### Clustered Longitudinal
Finally, clustered longitudinal data combines elements of both *longitudinal* and *clustered* data. In these types of datasets, measurements are taken repeatedly from the unit of analysis, but these units are also grouped within clusters. We therefore have a correlation structure that relates to both the repeated measurements over time and to the clusters. 

As an example, we will look at a dataset where measurements of gingivitis were taken from teeth that had been fitted with dental veneers. These measurements were taken 3 months after fitting and 6 months after fitting. This time gap is long and has a clear temporal ordering, so we would think of this as *longitudinal* rather than *repeated measurements*. In addition, *multiple* teeth were sampled from each patient. So, each tooth has two measurements over time and is clustered within a specific individual. This adds an addition layer of variation, because teeth from the same subject are likely to be correlated. Hence, this is *clustered-longitudinal* data.

````{admonition} A Note on Time Series Data
:class: tip
As a point of clarity, another type of data you may come across is *time series* data. Unlike longitudinal data, time series data are generated from a *single* unit and tend to have many more measurements taken much closer in time. One example could be real-time measurements of gait from someone with Parkinson's disease who is asked to perform a walking task. By the end of the task, there may be hundreds or thousands of measurements taken from across time on this single unit. Importantly, these type of data will exhibit a *very specific* correlation pattern, where values close together in time will be *more correlated* than values further away in time. This requires a specialised correlation structure to capture, known as an *autoregressive* model. This parameterises the correlation as decreasing as a function of time (e.g. an AR(1) model). These types of structure are more difficult to specify using mixed-effects models and provide a good opportunity to use GLS.
````

## Structures as Hierarchies
Now we have discussed the different types of dataset, it is worth remembering that our recommended approach for building a mixed-effect model is to start by *thinking* in multilevel/hierarchical terms and then collapsing these to mixed-effects form. As such, we also want to think of all these datasets in terms of *levels of variation* that we can directly translate into a multilevel model. The table below, taken from [West, Welch & Galecki (2022)](https://www.taylorfrancis.com/books/mono/10.1201/9781003181064/linear-mixed-models-brady-west-kathleen-welch-andrzej-galecki?context=ubx&refId=4dbda113-7f59-4ba9-a710-897ef536cfee), breaks each data type and dataset down into multilevel form. For each example, the *unit* of analysis is given in *italics*. This is an aspect we will define more formally a little further below.


| **Data Type**    | **Clustered (2-level)** | **Clustered (3-level)** | **Repeated Measures**              | **Longitudinal**      | **Clustered Longitudinal** |
|------------------|-------------------------|-------------------------|------------------------------------|-----------------------|----------------------------|
| **Example**      | Rat Pup                 | Classroom               | Rat Brain                          | Autism                | Dental Veneer              |
| **Level 1**      | _Rat Pup_               | _Student_               | Replications within each treatment | Measures across years | Measures across months     |
| **Level 2**      | Litter                  | Classroom               | _Rat_                              | _Child_               | _Tooth_                    |
| **Level 3**      |                         | School                  |                                    |                       | Patient                    |

The easiest way to think about this is as follows:

- **Level 1** represents observations at the *most detailed* level of the dataset. For *clustered* data, this represents the *units* of analysis, whereas for *repeated measurement/longitudinal* data this represents the repeated measurements collected for each unit. Remember, the outcome variable is always measured at Level 1 and so Level 1 represents the finest granularity of whatever that measurement is.
- **Level 2** represents the next level in the hierarchy. For *clustered* data, this will be the *clusters of units*. For *repeated measurement/longitudinal* data, this will represent the *units* of analysis. 
- **Level 3** represents the next level in the hierarchy, above Level 2. For *clustered* data, this will represent *clusters of clusters*. For *clustered longitudinal* data, this will represent clusters of the units defined at Level 2.

There are not usually more than 3 levels of variation within these types of datasets, though this is neither a rule nor a limitation. Part of the challenge of using a multilevel conceptualisations is defining what a particular dataset actually is. We can use the rules above to define each level of the data, but only if we know what type of structure we have. To facilitate this, we will now turn to a set of steps you can use to try and determine the structure you have. Once this is known, apply the definitions of the levels above and you should be able to write a table similar to the above for your data. This will be the first step in building the most suitable model. 

## How to Determine the Structure of a Dataset?
As a final section in this part of the lesson, we present a simple 3-step approach to determining the data structure. Although reasonably generic, this should be widely applicable and should always help you to at least *think* more clearly about any dataset you are working with. Even if the structure of the *whole* dataset does not easily fit, it is entirely possible that some subset *would* fit into one of the definitions above. So it is always worth thinking flexibly, especially if you have a very large dataset at your disposal. It is also worth knowing that these steps can be applied to *any* dataset, not just those with some sort of hierarchical structure. This can therefore be considered the starting point for *any* data analysis.

## Step 1. Determine the *unit* of analysis
As a first step, you need to determine your *unit of analysis*. Broadly speaking, this is the *entity whose outcome the model attempts to explain or predict*. For instance, in the case of repeated measurements, this is almost always *subject*. In the case of the clustered school example, the unit of analysis is *student*, because it is the student's maths scores that we are attempting to explain. Similarly, in the clustered-longitudinal dental veneer example, the unit of analysis is *tooth* because interest lies in measurements from the teeth themselves. 

Another way to think about this is what statements do we want the model to make? What is our model *about*? Is it about the maths scores of classes and schools, or is it about the maths scores of *students*? Is it about the weight of rat litters, or is it about the *rat pups* themselves? Is it about measures of gingivitis in a given month, measures of gingivitis from a specific patient, or measures of gingivitis from *teeth*? You need to ask yourself: "If I estimated an effect in this model, what entity would it describe?”. If the model predicts *weight*, what is this the weight of? If the model predicts maths scores, who's maths scores are they? If the model predicts gingivitis over time, what entity doo these measurements correspond to?

## Step 2. Determine what each *row* of the dataset represents
As a next step, we need to determine what each row of the dataset corresponds to. Importantly, this assumes that data is in *long format*, so any wide-formatted data need to be converted first. Once the data is suitably organised, we have two possibilities

1. If each row is *one unit* measured *multiple times*, then this is either repeated measures or longitudinal.
2. If each row is *one unit* measured *once* then either there is no dependency structure, *or* the units may be arranged in *clusters*.

````{admonition} Data Organisation
On a practical note, you may need to reorganise the data by the unit of analysis, so that the structure is not missed. For instance, if the dataset contains a `subject` index, if may not necessarily be organised by subject. This means that the repeated measurements for each subject are spread across the dataset, rather than being organised together. This can make it easy to miss the structure, especially if you are just calling something like `head(data)` to look at the first few rows.
````

## Step 3. Determine whether you have clusters
As a final step, we need to determine whether there are any *clustering* structures. In terms of identifying clustering variables, we need to think carefully because it can be easy to mix up a *grouping variable* and a *clustering variable*. In principle, a *grouping variable* (such as experimental treatment, diagnosis or sex) is something that each unit *HAS*, with only a fixed number of levels that we are interested in. By comparison, a *clustering variable* (such as litter, clinic or school) is something that each unit is *IN*, with a selection of levels representing a random sample of all possible levels.

We can use a *visual metaphor* here to try and make the distinction clear: *rubber bands* vs *sticky notes*. Clusters are *rubber bands* that tie certain units together due to a shared context. Each rubber band is only one possibility from a range of other possibilities. For instance, we only sample a small number of possible rat litters, or schools or clinics. By comparison, *grouping variables* are *sticky notes* that label a characteristic of each unit. For instance, we can label which treatment each unit received, or which diagnosis they have been given. These are not random samples from all possible levels of treatment or diagnosis, rather they correspond to an exhaustive set of labels within this experiment.

Going back to the example datasets, rat pups are clustered within *litters*. "Litter" is not a characteristic of an individual rat pup, it is something the pup is *IN* rather than something the pup *HAS*. The litters in the experiment are a random sample of all possible rat litters, rather than an exhaustive set of experimental manipulations. The differences between the litters is not a systematic effect that we are interested in, rather each litter represents a *shared latent influence*. Similarly, students are clustered within *classes*, which are further clustered within *schools*. The classes are a random sample of all possible maths classes and the schools are a random sample of all possible schools. Each student is *IN* a class and *IN* a school, rather than the class or school being something the student *HAS*. Differences between classes and schools are not taken as a systematic effect of interest, rather they are taken as indicative of *shared latent influences* (same teachers, same school environment, same classmates, same local area etc.) that make some units more similar to each other than others.

````{admonition} Checking for Multiple Structures
It is also important during these steps to not assume that there is only a *single* structure. If we have found one type of structure, we need to check carefully to see if there are any others. For instance, if we have identified clusters of units, we then need to check and see if there are *clusters* of clusters. Similarly, if we have identified repeated measurements, we need to check and see if there are also *clusters* or perhaps *longitudinal* measurements as well. For instance, repeated measurements are taken within several different time-points.
````

[^ratbrain-foot]: Which would be rather hard for a rat to do while its head is cut open, but we digress...
