# Beyond Repeated Measures

... The advantage is that any covariance structure, no matter how complex, will be *automatically* taken into consideration by the mixed-effects model. The only caveat is that we have to express the structure within the data correctly in order for this to work. So, the focus of mixed-effects models is capturing the *structure* of the data, with the correlation falling-out as a byproduct. This means that we can use mixed-effects models to capture all sorts of complex structures, *beyond* repeated measurements. We do not even need to think about how these structures translate into complicated variance-covariance matrices, we just let the model work it out.
