# Introduction
This lesson represents the end of the mixed-effects content on this unit. Last week, we introduced the mixed-effects framework, but made allusions to the fact that our example did not leverage the full power of this approach. In addition, throughout this section of the unit, our focus has been on mixed-effects in the context of *repeated measurements*. Although this is typically where dependence would be found within psychology and cognitive neuroscience, it is important to know that mixed-effects models are much *broader* than just repeated measurements. Indeed, the power of mixed-effects approaches is their ability to represent a *variety* of complex data structures, of which repeated measurements are just one possibility. 

For both these reasons, this final lesson will broaden your perspective on mixed-effects models, introducing various different types of data structure that mixed-effects models can be applied to. These include *clustered*, *repeated measurements*, *longitudinal* and *clustered longitudinal* data. As part of this, we will discuss how to identify and classify your own data. Although these categories are not necessarily exhaustive, you should always be able to define your data as *close* to one of these and thus should always have some sort of starting point.

In addition, this lessons aims to leave you with a practical set of guidelines for actually *building* a mixed-effects model from first principles. The aim is to provide guidelines that can be applied to a variety of datasets, allowing you to build complex models layer-by-layer and, most importantly, understand what the model is saying at each step. The bulk of this lesson will therefore be dedicated to a single example of this process. This contains a lot more detail than examples typically do, precisely because you need to understand each step as deeply as possible. To support this, examples of the same process applied to *different* data structures will be provided as part of the associated workshop. These can be considered a set of *templates* that you can use for your own analyses.
