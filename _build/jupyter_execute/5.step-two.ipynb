# Building the Model: Step 2




````{admonition} Aim: Expand the Model to Multiple Dependency Structures
:class: info
The overall aim of Step 2 is to expand the model from Step 1 to *multiple* dependency structures. For instance, *multiple* clusters or *multiple* subjects. In doing so, we characterise which of the parameters from Step 1 we imagine *change* with each new dependency structure and which of the parameters stay *constant*. This is effectively a decision about which terms are *random effects* and which are *fixed effects*.
````

For step 3, we need to imagine that we had *multiple* litters and then need to rewrite the model above accordingly. To begin, we need another index to make it clear *which litter* the model is referring to. To do this, we will use $k$ to indicate `litter`, but will use a slightly different notation to illustrate which elements *belong* to the model for litter $k$. So, we write the model for litter $k$ as

$$
y^{(k)}_{ij} = \mu^{(k)} + \alpha^{(k)}_{j} + \eta^{(k)}_{ij}
$$

This indicates that *everything* in the model depends upon the specific litter. This is generally our theoretical starting-point. However, in the next step, we will determine whether everything here *can* change within each litter, given the data we have, and whether we *want* them to change with each litter. We have already gone some way towards this in the previous step by fitting the model for a single litter, but we will formalise it now.

## Identifying Which Terms are *Random*
Now that we have our model for each level of `litter`, we need to make some decisions. As indicated above, every element has the *potential* to be different for each litter. Of course, the data itself always will be, as will the errors. So our question is about how to treat those elements that form the *mean function* of each litter:

$$
E\left(y^{(k)}_{ij}\right) = \mu^{(k)} + \alpha^{(k)}_{j}
$$

For each of these term, we have a *choice*. Do we imagine that this effect changes with each litter, or do we want to treat it as a *global* effect that is assumed to be the *same* across all litters? In other words, we are trying to capture *why* the `litter` variable matters. What influence does `litter` have on the weight of the pups within it? Any element that *changes* from litter-to-litter captures why `litter` influences `weight`, whereas any element that is the *same* suggests that `litter` is irrelevant. 

As will hopefully become clear, there is no *absolute truth* here. These decisions have to be made by reasoning about the problem and considering what the model is saying, given the choices we make. Although we can use model comparisons to see which choice is most compatible with the data, we also need to start from a position of justifying our choices. So there is no shortcut to engaging your brain and thinking about the problem in front of you. This can be difficult, but is part of the reality of analysing data and part of your worth as a data analyst. 

#### How to Treat $\mu^{(k)}$?
To start with, let us consider how to treat $\mu^{(k)}$. This is the *average weight* per-litter. So, from a purely *theoretical* perspective, do we think that each litter should get its own baseline weight, or do we want to constrain the model so that all litters share the same baseline weight? In other words, do we think that the reason that `litter` is important is because it influences the average weight of the pups within each litter, or do we think that each litter has the same baseline weight and any deviations are sampling noise that is unrelated to litter? For this model, we choose to allow the baseline weight to vary and so our model per-litter remains

$$
E\left(y^{(k)}_{ij}\right) = \mu^{(k)} + \alpha^{(k)}_{j}
$$

The reasoning behind this is explored in the drop-down below

````{admonition} Reasoning About the Intercept
:class: info, dropdown

Our choice in terms of the intercept $\mu^{(k)}$ is effectively a decision between the following models. Either we allow $\mu$ to change from litter-to-litter 

$$
E\left(y^{(k)}_{ij}\right) = \mu^{(k)} + \alpha^{(k)}_{j}
$$

or we fix $\mu$ to the same value for each litter

$$
E\left(y^{(k)}_{ij}\right) = \mu + \alpha^{(k)}_{j}.
$$

To think about this more clearly, imagine that we were able to take *two* mother rats and make them give birth to new litters over and over again. Our key question is whether we believe that 

1. Over time the average weight of all the litters, irrespective of their mother and rearing environment, will converge on the *same value*. In other words, the mother and rearing environment *does not matter*. Any discrepancy in our measurements is simply sampling variation that is unrelated to each mother and rearing environment.
2. Over time, the average weight of all litters from the same mother will converge on their own *unique* value that will be different from another mother. In other words, the mother and rearing environment *does* matter. For instance, mother rat A always produces litters that are heavier on average than mother rat B. Any discrepancy in our measurements is then a mixture of sampling variation *and* the effect of a specific mother and rearing environment.

This is precisely the choice between treating the mean weight per-litter as a *constant* across litters or as unique *per-litter*. Our question is which of these descriptions of reality is more plausible?

In this example, it is probably not sensible to constrain all litters to the same expected weight. Depending upon different genetic factors and birthing conditions, some litters will be heavier or lighter than others. This goes beyond just random sampling noise and becomes a systematic source of variation that directly relates to the different litters. In fact, this was the whole reason for characterising `litter` as a clustering variable in the first place. So, in this instance, we will argue that it makes most sense to allow $\mu^{(k)}$ to differ per-litter. Irrespective of whether we are dealing with *clusters* or *subjects*, it is *almost always* sensible to start with this *random-intercepts* model.
````

Once we have made this decision, we also need to examine the data from a *practical* perspective, to see if it can support a different average value of `weight` per-litter. Even if we *want* to do this, that data needs to allow it. This requires *at least* two values of `weight` for each litter, so that there is variance left over once $\mu^{(k)}$ has been fit. In this example, each litter has multiple values and so a varying-intercept is perfectly possible. So, we conclude that both *theoretically* and *practically* we can allow a different baseline weight per-litter.   

#### How to Treat $\alpha^{(k)}_{j}$?
Now, we need to provide the same consideration to $\alpha^{(k)}_{j}$. Remember, this is the *effect* of `sex` on `weight` *per-litter*. Unlike the random-intercepts, the decision about how to treat $\alpha^{(k)}_{j}$ is less clear-cut. If our bare minimum is include an intercept per-litter, then we can still capture the importance of litter whether we also allow $\alpha^{(k)}_{j}$ to vary or not. So this choice is much more about our *reasoning* than it is about the need to capture `litter` as a clustering variable. For the purpose of this analysis, we will start by assuming that the effect of `sex` can differ per-litter and so our model remains

$$
E\left(y^{(k)}_{ij}\right) = \mu^{(k)} + \alpha^{(k)}_{j}
$$

This is again explored more in the drop-down below

````{admonition} Reasoning About the Slope
:class: info, dropdown
As with the intercept parameter, we will consider our two modelling options. If we allow the effect of `sex` to differ per-litter, we are assuming

$$
y^{(k)}_{ij} = \mu^{(k)} + \alpha^{(k)}_{j} + \epsilon^{(k)}_{ij}
$$

Alternatively, if we fix the effect of `sex` to be the *same* across all litters, we are assuming

$$
y^{(k)}_{ij} = \mu^{(k)} + \alpha_{j} + \epsilon^{(k)}_{ij}.
$$

So, our question is one of *reasoning*. Do we think that there is a universal effect of `sex` on `weight` that is always there irrespective of genetics or rearing environment? Or, do we believe that the magnitude of the effect of `sex` on `weight` depends, at least to some degree, on both genetics and rearing environment? 

Of course, the problem here is that you could argue it *either way*. It very much depends on how strongly we believe that the effect of `sex` on `weight` is homogeneous across litters, or how strongly we believe that there could be a gene $\times$ environment interaction that change the effect of `sex` for each litter? Is it plausible that some mother rats may consistently have female pups that are *heavier* than male pups? Or female and male pups that are largely *equivalent* in weight? Or, is it more plausible to think that male pups will *always* be heavier than female pups on average, irrespective of mother or rearing environment?

When faced with this sort of decision, we either have to (a) pick one possibility that we believe more strongly or (b) pick the most general case and then use model comparisons to see which the data supports. If you are working alongside domain experts on the analysis, this is the sort of conversation you could have with them to pick the option that is most plausible. In the absence of this, we will take the *second* approach, which is to temporarily assume the more general case of the `sex` effect differing per-litter, but with a view to using model comparisons for the final decision on this[^sextextbook-foot].
````

Once we have made this decision, we need to examine the data from a *practical* perspective, to see if it can support a different `sex` effect per-litter. In order to do so, we need *at least* two male pups and two female pups on average per-litter. If each litter only consisted of a single male and a single female, this would not be possible. Similarly, if each litter consisted of only a single sex, this would also not be possible. We can have a look at the average number of male and female pups across litters using

`````r
colMeans(xtabs(~ litter + sex, data=ratpup))
`````

This indicates that there should be enough variation on average per-litter to allow this. Even if there was not, we would soon find our because the model would fail to fit. However, we can at least avoid this embarrassment by trying to reason about it first. Although tempting to brute-force these decisions by examining when the model fails, this is not a good strategy because it suggests that you do not *understand* the model, nor why it is failing to fit. You should have a clear idea abut this *before* even approaching the software.
