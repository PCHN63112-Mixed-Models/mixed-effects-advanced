## The General Steps
... The assumption is that you have identified the data structure as a *prerequisite* to building the model.

1. Model a single dependency structure
    - Repeated measurements/longitudinal: write the model for *one subject*
    - Clustered: write the model for *one cluster*
    - Clustered longitudinal: write the model for *one subject* within *one cluster*
2. Add multiple dependency structures
3. Write the higher level models

A final step, if you are using *mixed-effects* software, rather than *multilevel/hierarchical* software, is to collapse the model across the levels to produce a single expression.

## Step I 

The overall aim of Step 1 is to write the model we want for only a *single* dependency structure. So, *one* cluster, or *one* subject. This needs to be at the *lowest* level of the hierarchy, so no other dependency structures should exists *within* this. For instance, if you had *clustered longitudinal* data, you would select *one subject* within *one cluster*. You would not select a cluster as a whole, because that would contain multiple subjects. We are trying to select the smallest single block of dependency that we can.

````{admonition} Definition: Dependency Structure
:class: info
A set of observations that are statistically dependent because they share the same source of random variation.

| Design                 | Dependency structure                               |
| ---------------------- | -------------------------------------------------- |
| Repeated measures      | All observations from one subject                  |
| Clustered data         | All observations from one cluster                  |
| Clustered longitudinal | All observations from one subject (within cluster) |

A dependency structure is a bundle of observations that *belong together* in some fashion.
````

## Step II
The overall aim of Step 2 is to expand the model from Step 1 to *multiple* dependency structures. For instance, *multiple* clusters or *multiple* subjects. In doing so, we characterise which of the parameters from Step 1 we imagine *change* with each new dependency structure and which of the parameters stay *constant*. This is effectively a decision about which terms are *random effects* and which are *fixed effects*.

## Step III
The overall aim of Step 3 is to write the models that explain the terms discussed in Step 2. These models will be slightly different depending upon whether these terms are *fixed* or *random*. 

## Examples of Building Mixed-effects Models
For the remainder of this lesson, we will work slowly through a *single* example of building a mixed-effects model. This is quite detailed and may feel a bit overwhelming, particularly in terms of imagining being able to do the same thing with your own data. To help, further examples of this process will be given as part of the workshop this week. These will run through the same steps, but on different types of data and can be treated as *templates* for your own analyses. Once the structure of your own data has been determined, you can use the *closest* example as a starting point. Eventually, you will get the hang of doing this and how these models can be built slowly from the ground-up. Remember as well, this approach is very purposeful in trying to help you understand these models more fully than applied researchers often do. Taking your time to build them up layer-by-layer allows you to fully absorb everything the model is saying and the choices you are making. In the long-run, this is much more effective than using templates of `R` syntax to specify a model that you do not fully grasp. Remember, if you ever see a researcher just mechanically adding `1|id` or `1|subject` to their model and leaving it there then there is a good chance they they do not fully grasp what they are doing.
