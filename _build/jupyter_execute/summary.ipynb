# Summary
This lesson represents then end of our journey into the world of mixed effects. The main aims of this lesson were twofold. Firstly, conveying the idea that mixed-effects are much more general than repeated measurements and can be applied to a variety of data structures, whether they contain repeated measurements or not. Secondly, demonstrating the practical process of building these models slowly from the ground-up. As a consequence, there was a lot of information here and it would be unrealistic to assume that it can all be absorbed in a single reading. You will need to come back to this many times before a full sense of these models can be obtained. This is particularly true in terms of actually building these models. It is one thing to *read* the process, but something else to try and do it yourself. This is where *practice* is important, as well as having a corpus of examples available to support this practice. In the workshop this week, we will have *multiple* examples of this process applied to a variety of examples. Once you see this more and start trying to do it yourself, it should become less alien. Just remember, this will take time and practice to achieve.
