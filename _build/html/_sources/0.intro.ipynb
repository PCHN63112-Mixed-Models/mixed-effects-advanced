# Introduction
In this lesson... Throughout this section of the unit, our focus has been on mixed-effects in the context of *repeated measurements*. Although this is typically where dependence would be found within psychology and cognitive neuroscience, it is important to know that mixed-effects models are much *broader* than just repeated measurements. Indeed, the power of mixed-effects approaches is their ability to represent complex data structures, of which repeated measurements are just a single possibility. ...
