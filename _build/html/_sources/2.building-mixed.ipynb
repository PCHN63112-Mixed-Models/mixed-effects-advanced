# Building Mixed-effects Models
In this part of the lesson, we will cover some generic guidance on building mixed-effects models. The assumption is that you have identified the structure of your data and will be using this to guide how to build the model. We can then follow 3 general steps to build each level slowly, before (optionally) collapsing the levels to mixed-effects form. The discussion here will necessarily be quite general. After this point, the remainder of the lesson will be dedicated to applying these steps in detail to a single example.

## The General Steps
To begin with, the general steps of building the model are as follows:

1. Model a single dependency structure
2. Add multiple dependency structures
3. Write the higher level models

Although there are several complexities within each of these steps, they provide a very general-purpose and logical sequence for creating a model from first principles. After completing these steps, you may need to collapse the model across the levels to produce a single mixed-effects expression. This is only necessary if you are using *mixed-effects* software rather than *multilevel/hierarchical* software, or if you prefer thinking about these models in mixed-effects form.

We will now discuss each step in detail, before seeing them applied to a single dataset later in the lesson.

## Step I in Detail 
The overall aim of Step I is to write the model we want for a *single* dependency structure. For instance, the model for *one* cluster or for *one* subject. This needs to be at the *lowest* level of the hierarchy, so no other dependency structures should exists *within* this. For instance, if you had *clustered longitudinal* data, you would select *one subject* within *one cluster*. You would not select a cluster as a whole, because that would contain multiple subjects. In effect, we are trying to select the smallest chunk of dependency that we can. If we end up with only a single row of our dataset, then there are *no* dependencies and we can go back to the normal linear model.

This idea of a *dependency structure* is explored more in the box below. In effect, we are trying to define the *boundaries of dependency*. If we think in terms of block-diagonal covariance matrices, we need to identify what the smallest *block* would look like. In repeated measurements, each subject has a covariance block. In clustered data, each cluster has a covariance block. When there are multiple levels, we have blocks within blocks (e.g. classes within schools). In each case, we just need to find the *smallest* block we can work with and focus only on modelling that block. We then pretend that one such block is the only data we have and build the model from there.

````{admonition} Definition: Dependency Structure
:class: info
A more formal definition of a *dependency structure* is a set of observations that are statistically dependent because they share the same source of random variation. Depending on the type of data you are working with, these structures will be associated with different entities:

| Design                 | Dependency structure                                   |
| ---------------------- | ------------------------------------------------------ |
| Repeated measures      | All observations from one subject                      |
| Clustered data         | All observations from one cluster                      |
| Clustered longitudinal | All observations from one subject (within one cluster) |

</br>

Note that the *unit of analysis* can sit in different positions relative to the dependency structure. In repeated measurements, the unit of analysis *is* the dependency structure. In clustered data, multiple units sit *within* the dependency structure. In the rat pup example, *litter* is the dependency structure, with multiple pups sitting *within* a litter. In the longitudinal autism example, each *child* is the dependency structure, with longitudinal measurements taken from within each child. 

Looking back at the table from the previous part of this lesson, we can see that these structures as a whole live at Level 2, with individual examples living at Level 1. So we have rat pups (Level 1) from *within* litters (Level 2). Students (Level 1) from *within* classrooms (Level 2). Repeated measurements (Level 1) from *within* rat brains (Level 2). Longitudinal measurements (Level 1) from *within* a tooth (Level 2).
````

## Step II in Detail
The overall aim of Step II is to expand the model from Step I to *multiple* dependency structures. For instance, *multiple* clusters or *multiple* subjects. So, we take our single model of one dependency structure, and imagine that we now have many clusters or many subjects. The key question is deciding which of the parameters of the model from Step I will change with each new structure. This is effectively asking which of the effects we think will differ from *subject-to-subject* or from *cluster-to-cluster*. Importantly, this needs to be done thinking about the *population* model. If we had the whole population available, do we imagine that all estimates would converge on a single value, or do we think that there would still be some fundamental difference between different subjects or between different clusters? 

An example can make this clearer. In the rat pup dataset, we may choose to model a single litter using the effect of *sex* on weight. Once we introduce multiple litters, we have a choice. Do we imagine that the effect of sex on weight is *different* for each litter, or is it the same? Again, we are thinking at the level of the *population* here, not estimates from individual samples. So, our choice is about whether we treat the effect of sex on weight as a singular fact of biology that is true irrespective of the specific litter, or do we imagine that there is something about the shared biology and rearing environment of a single litter that can *change* how sex influences weight? If we had an infinite number of litters containing an infinite number of pups, would the effect of sex converge to the same value within each litter, or would there still be differences? Do the litters themselves influence the effect of sex on weight, or is the specific litter irrelevant?

Importantly, this is the part of the model that can be the *most difficult* to reason about and there are often multiple plausible options. In the example above, there could be arguments either way about how to treat the effect of sex on weight. In general, we recommend making the most sensible choice that you can. However, you can also take a much broader perspective and start with the assumption that *all* effects that *can* vary across dependency structures *do* vary. At this point, the question is simply whether the data that you have supports this type of model, rather than whether it makes sense in context. You can then use model comparisons to see which of these assumptions is better supported and build the final model from there. 

## Step III in Detail
The overall aim of Step III is to write the models that explain the terms discussed in Step 2. These models will be slightly different depending upon whether these terms are *random* (allowed to vary across dependency structures) or *fixed* (treated as the same, irrespective of the dependency structure). As we will come to see, we generally index these terms per-structure (e.g. include a *subject* index or a *cluster* index), but depending on whether they are assumed fixed or random will determine whether these higher-level models also include *random error*. This random error allows each structure to deviate individually from the population mean. If we do *not* include this, each structure is assumed to not individually deviate and the effect is fixed. Any deviation that we see within a particular sample is just sampling error, rather than a systematic deviation related to the structure itself. 

As an example, if *litter* is assumed *random* then each litter is allowed to deviate from the population mean weight and thus form its own mean weight. This captures the idea that each litter is a random draw from a population of litters. Alternatively, if litter is assumed *fixed*, then we assume that each litter shares the same mean weight and any deviation we see is purely due to sampling error and nothing else. Visually, this is the difference between assuming that each litter has its own distribution whose mean is taken from a larger distribution of litter means, and assuming that each pup's weight is just taken from a global distribution of weights where litter does not matter. The *random* view is that we are working with *two* distributions, whereas the *fixed* view is that we are working with *one* distribution. Admittedly, this all sounds a bit vague and conceptual at present, but will hopefully become more concrete once you have seen further examples. 

This step is also the point where we can introduce *grouping variables*. As these variables tend to correspond to manipulations that are applied *as a whole* to different dependency structures, they can be introduced once multiple dependency structures are represented in the model. For instance, once multiple *litters* are available, we can include an effect of *treatment*, given that this was applied to litters as a whole. For this reason, the higher-level models can be thought of as the *between-subjects* structure of the data, though this is mainly applicable to repeated measurements. 

## Examples of Building Mixed-effects Models
For the remainder of this lesson, we will work slowly through a *single* example of building a mixed-effects model. This is quite detailed and may feel a bit overwhelming, particularly if you start trying to imagine doing this yourself with any arbitrary dataset. To help, further examples of this process will be given as part of the workshop this week. These will run through the same steps on different types of data and can be treated as *templates* for your own analyses. Once the structure of your own data has been determined, you can use the *closest* example as a starting point. Eventually, you will get the hang of doing this and how these models can be built slowly from the ground-up. 

Remember as well, this approach is very purposeful in trying to help you understand these models more fully than applied researchers often do. Taking your time to build them up layer-by-layer allows you to fully absorb everything the model is saying and the choices you are making. In the long-run, this is much more effective than using templates of `R` syntax to specify a model that you do not fully grasp. With this in mind, if you ever see a researcher just blindly adding `1|id` or `1|subject` to their model and leaving it there, then there is a good chance they they do not fully grasp what they are doing. We want you to be better than this, which is why we are spending so much time here.
